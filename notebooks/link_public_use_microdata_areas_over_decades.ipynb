```python
# Ensure states and pumas lists are of the same length
assert len(states) == len(pumas), "States and PUMAs lists must have the same length"

# Create a list of (state, puma) tuples
state_puma_pairs = list(zip(states, pumas))

# Prepare an empty DataFrame to store matched rows
gdf_matched_unique = []

# Iterate over each (state, puma) pair
for state, puma in state_puma_pairs:
    # Filter gdf for the current pair
    temp_df = gdf[(gdf['STATEFP20'] == state) & (gdf['PUMACE20'] == puma)]
    
    # If multiple rows match, select a row based on a criterion
    # Here, we simply take the first occurrence using `.iloc[0]`
    # Note: You should define the selection criterion based on your data's context
    if not temp_df.empty:
        gdf_matched_unique.append(temp_df)

# Reset index of the final DataFrame
gdf_matched_unique = pd.concat(gdf_matched_unique).reset_index(drop=True)
```

```python
%%sql
COPY(
WITH puma_2010 AS (
  SELECT * FROM '/Users/me/data/american_community_survey/PUMA5/2010/combined/TIGER2010_PUMA10.shp'
),
us_cbsa_2020 AS (
  SELECT * FROM '~/data/american_community_survey/shapefiles/tl_2020_us_cbsa.shp'
),
new_york_cbsa_2020 AS (
  SELECT * FROM us_cbsa_2020
  WHERE NAMELSAD LIKE '%New York%'
),
intersecting_rows AS (
  SELECT puma.*
  FROM puma_2010 puma
  JOIN new_york_cbsa_2020 nycbsa
  ON ST_Intersects(puma.geom, nycbsa.geom)
)
SELECT * FROM intersecting_rows
) TO '~/data/american_community_survey/shapefiles/newyork_newark_cbsa_2010_pumas_intersection_with_2020_cbsa.shp' WITH (FORMAT GDAL, DRIVER 'ESRI Shapefile');
```
